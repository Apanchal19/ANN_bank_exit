```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Importing the dataset
dataset = pd.read_csv(r"C:\Users\H P PANCHAL\Desktop\Data_Science\Bank_Customer_exit\Predict_customer_Will_Leave_the-Bank-master\BankCustomers.csv")
```

```python
# Making dependent and independent datasets based on features
x= dataset.iloc[:, 3:13]
y= dataset.iloc[:, 13]
```

```python
dataset.head()
```

```python
x.head()
```

```python
y.head()
```

```python
# convert categorical feature into dummy variables
states=pd.get_dummies(x['Geography'],drop_first=True)
gender=pd.get_dummies(x['Gender'],drop_first=True)
```

```python
states.head()
```

```python
gender.head()
```

```python
#drop the columns as it is no longer required
x=x.drop(['Geography','Gender'],axis=1)
```

```python
x.head()
```

```python
#concatenate the remaining dummies columns
x=pd.concat([x,states,gender],axis=1)
```

```python
x.head()
```

```python
import sklearn
```

```python
# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train= sc.fit_transform( x_train)
x_test = sc.transform(x_test)
```

```python
# Part 2 - Now let's make the ANN!


# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
x.head()
```

```python
# Initialising the ANN
classifier = Sequential()


# Adding the input layer and the first hidden layer
classifier.add(Dense(activation='relu',input_dim=11,units=6, kernel_initializer='uniform'))


# Adding the second hidden layer
classifier.add(Dense(activation='relu',units=6,kernel_initializer='uniform'))


# Adding the output layer
classifier.add(Dense(activation='sigmoid',units=1,kernel_initializer='uniform'))
```

```python
classifier.summary()
```

```python
# Compiling the ANN


classifier.compile(optimizer= 'adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Fitting the ANN to the Training set

classifier.fit(x_train,y_train,batch_size=10,epochs=5)
```

```python
# Part 3 - Making the predictions and evaluating the model

# Predicting the Test set results


y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
#Test results can be 0 or 1

y_pred = (y_pred > 0.5 )
```

```python
y_pred
```

```python
# Making the Confusion Matrix


from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
```
